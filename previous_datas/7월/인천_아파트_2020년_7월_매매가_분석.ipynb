```
ic_ap_50=ic_ap.sort_values(by='거래금액(만원)',ascending=False).head(50)
ic_ap_50.head()
```

```
ic_ap_b50=ic_ap.sort_values(by='거래금액(만원)',ascending=True).head(50)
ic_ap_b50.head()
```

```
ic_ap_50['주소']=ic_ap_50['구']+' '+ic_ap_50['동']+' '+ic_ap_50['도로명']
ic_ap_50.head()
```

```
ic_ap_b50['주소']=ic_ap_b50['구']+' '+ic_ap_b50['동']+' '+ic_ap_b50['도로명']
ic_ap_b50.head()
```

```
gmaps_key = "AIzaSyDzLFtZNl-A1dKuyfg9fUrU4vJhX49gfQo"
gmaps = googlemaps.Client(key=gmaps_key)
```

```
from tqdm import tqdm_notebook

lat = []
lng = []

for n in tqdm_notebook(ic_ap_50.index):
    tmp_add = str(ic_ap_50['주소'][n])
    tmp_map = gmaps.geocode(tmp_add)

    tmp_loc = tmp_map[0].get('geometry').get('location')
    lat.append(tmp_loc['lat'])
    lng.append(tmp_loc['lng'])

ic_ap_50['lat'] = lat
ic_ap_50['lng'] = lng
ic_ap_50
```

```
lat = []
lng = []

for n in tqdm_notebook(ic_ap_b50.index):
    tmp_add = str(ic_ap_b50['주소'][n])
    tmp_map = gmaps.geocode(tmp_add)

    tmp_loc = tmp_map[0].get('geometry').get('location')
    lat.append(tmp_loc['lat'])
    lng.append(tmp_loc['lng'])

ic_ap_b50['lat'] = lat
ic_ap_b50['lng'] = lng
ic_ap_b50
```

```
ic_ap_50.reset_index(inplace=True)
ic_ap_50.drop('index',1,inplace=True)
ic_ap_50
```

```
ic_ap_b50.reset_index(inplace=True)
ic_ap_b50.drop('index',1,inplace=True)
ic_ap_b50
```

```
map = folium.Map(location=[37.4563, 126.7052], zoom_start=11)

for n in ic_ap_50.index:
    if pd.notnull(ic_ap_50['lat'][n]):
        folium.CircleMarker([ic_ap_50['lat'][n], ic_ap_50['lng'][n]], 
                                  radius=10, color='#CD3181', 
                                  fill_color='#CD3181',
                                  fill=True).add_to(map)
    
for n in ic_ap_b50.index:
    if pd.notnull(ic_ap_b50['lat'][n]): 
        folium.CircleMarker([ic_ap_b50['lat'][n], ic_ap_b50['lng'][n]], 
                                  radius=10, color='#3186cc', 
                                  fill_color='#3186cc',
                                  fill=True).add_to(map)
        
map
```
