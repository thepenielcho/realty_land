```
import json
import folium
import googlemaps
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```
gg_ap_50=ap_gg.sort_values(by='거래금액(만원)',ascending=False).head(50)
gg_ap_50.head()
```

```
gg_ap_b50=ap_gg.sort_values(by='거래금액(만원)',ascending=True).head(50)
gg_ap_b50.head()
```

```
gg_ap_50['주소']=gg_ap_50['구']+' '+gg_ap_50['동']+' '+gg_ap_50['도로명']
gg_ap_50.head()
```

```
gg_ap_b50['주소']=gg_ap_b50['구']+' '+gg_ap_b50['동']+' '+gg_ap_b50['도로명']
gg_ap_b50.head()
```

```
gmaps_key = "AIzaSyDzLFtZNl-A1dKuyfg9fUrU4vJhX49gfQo"
gmaps = googlemaps.Client(key=gmaps_key)
```

```
from tqdm import tqdm_notebook

lat = []
lng = []

for n in tqdm_notebook(gg_ap_50.index):
    tmp_add = str(gg_ap_50['주소'][n])
    tmp_map = gmaps.geocode(tmp_add)

    tmp_loc = tmp_map[0].get('geometry').get('location')
    lat.append(tmp_loc['lat'])
    lng.append(tmp_loc['lng'])

gg_ap_50['lat'] = lat
gg_ap_50['lng'] = lng
gg_ap_50
```

```
lat = []
lng = []

for n in tqdm_notebook(gg_ap_b50.index):
    tmp_add = str(gg_ap_b50['주소'][n])
    tmp_map = gmaps.geocode(tmp_add)

    tmp_loc = tmp_map[0].get('geometry').get('location')
    lat.append(tmp_loc['lat'])
    lng.append(tmp_loc['lng'])

gg_ap_b50['lat'] = lat
gg_ap_b50['lng'] = lng
gg_ap_b50
```

```
gg_ap_50.reset_index(inplace=True)
gg_ap_50.drop('index',1,inplace=True)
gg_ap_50
```

```
gg_ap_b50.reset_index(inplace=True)
gg_ap_b50.drop('index',1,inplace=True)
gg_ap_b50
```

```
map = folium.Map(location=[37.567167, 127.190292], zoom_start=9)

for n in gg_ap_50.index:
    if pd.notnull(gg_ap_50['lat'][n]):
        folium.CircleMarker([gg_ap_50['lat'][n], gg_ap_50['lng'][n]], 
                                  radius=10, color='#CD3181', 
                                  fill_color='#CD3181',
                                  fill=True).add_to(map)
    
for n in gg_ap_b50.index:
    if pd.notnull(gg_ap_b50['lat'][n]): 
        folium.CircleMarker([gg_ap_b50['lat'][n], gg_ap_b50['lng'][n]], 
                                  radius=10, color='#3186cc', 
                                  fill_color='#3186cc',
                                  fill=True).add_to(map)
        
map
```
